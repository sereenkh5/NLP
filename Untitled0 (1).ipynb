```python
from textblob import TextBlob
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns

#     
def analyze_sentiments(texts):
    results = []
    for text in texts:
        polarity = TextBlob(text).sentiment.polarity
        if polarity > 0:
            sentiment = 'Positive'
        elif polarity < 0:
            sentiment = 'Negative'
        else:
            sentiment = 'Neutral'
        results.append({'text': text, 'polarity': polarity, 'sentiment': sentiment})
    return results

#     
def plot_pie_chart(results):
    counts = {
        'Positive': sum(r['sentiment'] == 'Positive' for r in results),
        'Negative': sum(r['sentiment'] == 'Negative' for r in results),
        'Neutral': sum(r['sentiment'] == 'Neutral' for r in results)
    }
    labels = counts.keys()
    sizes = counts.values()
    colors = ['#66b3ff', '#ff6666', '#ffcc99']

    plt.figure(figsize=(6,6))
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
    plt.title('Sentiment Distribution (Pie Chart)')
    plt.show()

#   
def plot_bar_chart(results):
    counts = {
        'Positive': sum(r['sentiment'] == 'Positive' for r in results),
        'Negative': sum(r['sentiment'] == 'Negative' for r in results),
        'Neutral': sum(r['sentiment'] == 'Neutral' for r in results)
    }
    sns.barplot(x=list(counts.keys()), y=list(counts.values()), palette=['#66b3ff', '#ff6666', '#ffcc99'])
    plt.title('Sentiment Distribution (Bar Chart)')
    plt.ylabel('Number of Sentences')
    plt.show()

#      
def plot_wordcloud(results, sentiment_type):
    text = " ".join(r['text'] for r in results if r['sentiment'] == sentiment_type)
    if text.strip() == "":
        print(f"No {sentiment_type} sentences to generate word cloud.")
        return
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10,5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud for {sentiment_type} Sentences')
    plt.show()

#  
texts = [
    "I love this product! It works perfectly.",
    "However, the delivery was late.",
    "Overall, I am satisfied with the purchase.",
    "The packaging could be better.",
    "I hate how slow the app is.",
    "Customer support was very helpful.",
    "This is just okay, nothing special.",
    "Terrible experience, will not buy again."
]

#  
results = analyze_sentiments(texts)

#  
for r in results:
    print(f"Text: {r['text']}\nPolarity: {r['polarity']:.2f} | Sentiment: {r['sentiment']}")
    print('-'*50)

#   
plot_pie_chart(results)
plot_bar_chart(results)

#      
plot_wordcloud(results, 'Positive')
plot_wordcloud(results, 'Negative')
```
